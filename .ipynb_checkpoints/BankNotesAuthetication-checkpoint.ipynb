```python
#preprocessing data for model building
#method in Pandas is used to display the last few rows of a DataFrame named 'data'. 
data.tail()
```

```python
X = data.drop('class', axis = 1)
y = data['class']
```

```python
# from sklearn.model_selection import train_test_split
#This function call essentially splits your data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 1)
```

```python
#StandardScaler is a preprocessing technique used in machine learning to standardize the features by removing the mean and scaling them to unit variance. 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Logistic Regression
#Logistic Regression is a supervised learning algorithm used for classification tasks. Despite its name, it's used for binary classification problems where the target variable has two possible outcomes.
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
#function is being used to instantiate a logistic regression model, specifying the solver as "liblinear" and setting the random state to 1 for reproducibility.
log_classifier = LogisticRegression(solver = "liblinear", random_state = 1)
log_classifier.fit(x_train, y_train)
```

```python
accuracies = cross_val_score(estimator = log_classifier, X = x_train, y = y_train, cv = 10)
```

```python
print("Accuracies :\n", accuracies)
```

```python
print("Mean Accuracy : ", accuracies.mean())
```

```python
log_classifier_pred = log_classifier.predict(x_test)
log_classifier_pred
```

```python
from sklearn import metrics
```

```python
cm = metrics.confusion_matrix(y_test, log_classifier_pred, labels = [0, 1])
data_cm = pd.DataFrame(cm, index = [i for i in [0,1]], columns = [i for i in ['predicted 0', 'predicted 1']])
plt.figure(figsize = (8,6))
sns.heatmap(data_cm, annot=True)
```

```python
#SVC
from sklearn.svm import SVC
```

```python
#importing the Support Vector Classifier (SVC) module from scikit-learn (sklearn). 
svc_classifier = SVC(kernel = 'linear')
svc_classifier.fit(x_train, y_train)
```

```python
accuracies = cross_val_score(estimator = svc_classifier, X = x_train, y = y_train, cv = 10)
```

```python
print("Accuracies :\n", accuracies)
```

```python
print("Mean Accuracy : ", accuracies.mean())
```

```python
svc_classifier_pred = svc_classifier.predict(x_test)
svc_classifier_pred
```

```python
#The confusion matrix is a table that describes the performance of a classification model by comparing actual and predicted classes.
cm = metrics.confusion_matrix(y_test, svc_classifier_pred, labels = [0, 1])
data_cm = pd.DataFrame(cm, index = [i for i in [0,1]], columns = [i for i in ['predicted 0', 'predicted 1']])
plt.figure(figsize = (8,6))
sns.heatmap(data_cm, annot=True)
```

```python
rbf_classifier = SVC(kernel = 'rbf')
rbf_classifier.fit(x_train, y_train)
rbf_accuracies = cross_val_score(estimator = rbf_classifier, X = x_train, y = y_train, cv = 10)
print("Accuracies :\n", accuracies)
print("\n")
print("Mean Accuracy : ", accuracies.mean())
```

```python
rbf_classifier_pred = rbf_classifier.predict(x_test)
rbf_classifier_pred
```

```python
cm = metrics.confusion_matrix(y_test, rbf_classifier_pred, labels = [0, 1])
data_cm = pd.DataFrame(cm, index = [i for i in [0,1]], columns = [i for i in ['predicted 0', 'predicted 1']])
plt.figure(figsize = (8,7))
sns.heatmap(data_cm,annot=True)
```

```python
# from sklearn.ensemble import RandomForestClassifier
rfc_classifier = RandomForestClassifier(n_estimators = 50, criterion = 'entropy', random_state = 0)
rfc_classifier.fit(x_train, y_train)
```

```python
rfc_accuracies = cross_val_score(estimator = rfc_classifier, X = x_train, y = y_train, cv = 10)
print("Accuracies :\n", accuracies)
print("\n")
print("Mean Accuracy : ", accuracies.mean())
```

```python
rfc_classifier_pred = rfc_classifier.predict(x_test)
rfc_classifier_pred
```

```python
cm = metrics.confusion_matrix(y_test, rfc_classifier_pred, labels = [0, 1])
data_cm = pd.DataFrame(cm, index = [i for i in [0,1]], columns = [i for i in ['predicted 0', 'predicted 1']])
plt.figure(figsize = (8,7))
sns.heatmap(data_cm,annot=True)
```

```python
#kNeighbors classification
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = { 'n_neighbors': [3,2,5,7,2], 'p': [2, 3], 'leaf_size': [3,2,5,7,2]}
grid = GridSearchCV(KNeighborsClassifier(), param_grid = param_grid)
grid.fit(x_train, y_train)
grid.best_params_
```

```python
KNN = KNeighborsClassifier(n_neighbors = 2, p = 2, leaf_size = 3)
```

```python
KNN.fit(x_train, y_train)
```

```python
KNN_prd = KNN.predict(x_test)
KNN_prd
```

```python
rfc_accuracies = cross_val_score(estimator = KNN, X = x_train, y = y_train, cv = 10)
print("Accuracies :\n", accuracies)
print("\n")
print("Mean Accuracy : ", accuracies.mean())
```

```python
cm = metrics.confusion_matrix(y_test, KNN_prd, labels = [0, 1])
data_cm = pd.DataFrame(cm, index = [i for i in [0,1]], columns = [i for i in ['predicted 0', 'predicted 1']])
plt.figure(figsize = (8,7))
sns.heatmap(data_cm,annot=True)
```
